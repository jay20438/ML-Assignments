```python
def fit(x, y,learning_rate,epochs):

    x, y = validate(x, y)
    weight = np.ones((x.shape[1], 1))  # inicialized predictors
    num_samples = x.shape[0]  # numbers of samples
    for i in range(epochs):
        weight =weight -  learning_rate * (1 / num_samples) * x.T.dot((x.dot(weight) - y))


    return weight

def predict(x,weight):
    
    x = np.append(np.ones((x.shape[0], 1)), x, axis=1)
    return np.dot(x, weight)

def score(x,y, weight):
    x,y = validate(x,y)
    yprime = np.dot(x, weight)
    msr = np.sqrt(np.mean((yprime - y) ** 2))
    
    return msr

def validate(x,y):

    return   np.append(np.ones((x.shape[0], 1)), x, axis=1),\
             np.array(y).reshape(y.shape[0], 1)
```

```python
def k_fold_cross_validation(k, x, y):
    '''
        Performs K Fold Valaidation on given dataset.
        Splitting given dataset to k subsets.
    '''
    
    total_size, number_features = x.shape
    subset_size = total_size // k
    # print(subset_size)
    # print(x.shape,y.shape)
    
    # Store subsets - 
    x_subsets = []
    y_subsets = []
    
    # Splits the dataset to k subsets and stores them
    for i in range(k):
        x_slice = x[i * subset_size : min((i + 1) * (subset_size), total_size)]
        y_slice = y[i * subset_size : min((i + 1) * (subset_size), total_size)]
        
        # print(i * subset_size, min((i + 1) * (subset_size), total_size))
        x_subsets.append(x_slice)
        y_subsets.append(y_slice)
    
    # print(len(x_subsets[0]))
    # print(len(y_subsets[0]))
    
    # K Fold implemntation and calculate accuracy for each iteration and store
    rmse = []
    
    # Outer loop to select test set - 
    for i in range(k):
        x_test = x_subsets[i]
        y_test = y_subsets[i]
        x_train, y_train = np.empty((1, number_features)), np.empty((1, ))
        
        
        # Inner loop to select training set
        for j in range(k):
            if i != j:
                # print(x_train.shape, y_train.shape)
                # print(x_subsets[j].shape, y_subsets[j].shape)
                x_train = np.concatenate((x_train, x_subsets[j]), axis = 0)
                y_train = np.concatenate((y_train, y_subsets[j]), axis = 0)
        
        weights = fit(x_train,y_train,0.0000001,500)
        y_predicted = predict(x_test,weights)
        calc_score = score(x_test,y_test,weights)
        print("For fold: ", str(i), " k:", str(k))
        print("rmse: "+str(calc_score))
        rmse.append(calc_score)
        
    print("\nMean RMSE for "+ str(k)+" fold: " + str(np.mean(rmse)) + "\n")
        
for i in range(2, 6):
    k_fold_cross_validation(i, features, target)
```
